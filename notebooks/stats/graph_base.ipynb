```python
import pandas as pd

# Load the data to see its structure
file_path = '/Users/michael/Desktop/SpringData2024-main/data/regression/renewable_energy_difference.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

# Set the font
matplotlib.rcParams['font.family'] = 'Times New Roman'
matplotlib.rcParams['font.size'] = 15

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(18, 10))

# Define a color palette
palette = sns.color_palette("muted")

# Plot the renewable energy produced
sns.barplot(x="State", y="percentage_renewable_produced", data=data, color=palette[0], label="Produced", alpha=0.8)

# Plot the renewable energy consumed on top of the first plot
sns.barplot(x="State", y="percentage_renewable_consumed", data=data, color=palette[1], label="Consumed", alpha=0.8)

# Add a line plot for the difference
# We need a secondary axis for the difference as it's on a different scale
ax2 = ax.twinx()
sns.lineplot(x="State", y="difference", data=data, color=palette[2], marker="o", ax=ax2, label="Difference")

# Improve the legend
ax.legend(loc="upper left", frameon=True)
ax2.legend(loc="upper right", frameon=True)

ax.set_xlabel("State")
ax.set_ylabel("Percentage of Renewable Energy")
ax2.set_ylabel("Difference in Energy Production and Consumption")

plt.title("Comparison of Renewable Energy Produced vs. Consumed by State, Highlighting Difference")
plt.xticks(rotation=90)  # Rotate x-axis labels
plt.grid(True)  # Add grid
plt.tight_layout()  # Ensure everything fits
plt.show()
```
