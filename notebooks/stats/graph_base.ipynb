```python
import pandas as pd

# Load the data to see its structure
file_path = '/Users/michael/Desktop/SpringData2024-main/data/regression/renewable_energy_difference.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

# Set the font
matplotlib.rcParams['font.family'] = 'Times New Roman'
matplotlib.rcParams['font.size'] = 10

# Set the aesthetic style of the plots
sns.set_style("white")
sns.set(font="Times New Roman", context="notebook", font_scale=1.1)

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(18, 10), dpi=300)

# Define a color palette
palette = sns.color_palette("coolwarm")

# Plot the renewable energy produced
sns.barplot(x="State", y="percentage_renewable_produced", data=data, color=palette[0], label="Produced", alpha=0.6)

# Plot the renewable energy consumed on top of the first plot
sns.barplot(x="State", y="percentage_renewable_consumed", data=data, color=palette[3], label="Consumed", alpha=0.6)

# Change the color of the line plot for "difference" to green
ax2 = ax.twinx()
sns.lineplot(x="State", y="difference", data=data, color="green", marker="o", ax=ax2, label="Difference")

# Improve the legend
ax.legend(loc="upper left", frameon=True, fontsize=15)
ax2.legend(loc="upper right", frameon=True, fontsize=15)

# Set the labels and title with larger font size
ax.set_xlabel("State", fontsize=20)
ax.set_ylabel("Percentage of Renewable Energy", fontsize=20)
ax2.set_ylabel("Difference in Energy Production and Consumption", fontsize=20)
plt.title("Comparison of Renewable Energy Produced vs. Consumed by State", fontsize=20)

plt.xticks(rotation=90, fontsize=15)  # Rotate x-axis labels
plt.grid(True)  # Add grid
plt.tight_layout()  # Ensure everything fits
plt.show()
```
